```python
import numpy as np
import pandas as pd
import random
import math
import matplotlib.pyplot as plt
```

```python
class enterprise:
    def __init__(self,G,alpha,beta,
                 p = 5,std = 0.5,rho = 0.1, model = 'linear'):
        self.model = model
        self.G = G
        self.alpha = alpha
        self.beta = beta
        self.last_price = p
        self.std = std
        self.rho = rho
        self.price = [p]
        self.demand = []
        self.time = 1
        self.nash_price = 0
        self.nash_profit = 0
        self.check()
        
    def set_nash(self,price,profit):
        self.nash_price = price
        self.nash_profit = profit
    
    def get_time(self):
        return self.time
        
    def get_last_price(self):
        return self.last_price
    
    def return_price(self):
        return self.price
    
    def return_demand(self):
        return self.demand
        
    def check(self):
        print("OK")
    
    def update_time(self):
        self.time += 1
    
    def update_price(self,new = None):
        if new == None:
            self.price.append(self.price[-1]+self.rho*self.time**(-0.25))
            self.last_price = self.price[-1]
        else:
            self.price.append(new)
            self.last_price = new
            
    def update_demand(self,p_another):
        new = self.get_demand(p_another)
        self.demand.append(new)
        self.update_time()
                  
    def get_demand(self,p_another):
        var = random.normalvariate(0,self.std)
        temp = self.G - self.alpha*self.last_price + self.beta*p_another
        if self.model == 'linear':
            demand = temp
        elif self.model == 'exponential':
            demand = math.exp(temp)
        elif self.model == 'logit':
            demand = math.exp(temp)/(1+math.exp(temp))
        else:
            print("The demand model you input has to be 'linear','exponential' or 'logit'!")
        if demand+var < 0:
            return 0
        else:
            return demand+var
    
    def ols(self,competitor):
        c_price = competitor.return_price()
        now_price = c_price[-1]
        c_price = c_price[:-1]
        length = len(c_price)
        Y = np.array(self.demand).reshape((length,1))
        one = np.ones((length,1))
        x1 = np.matrix(self.price).reshape((length,1))
        x2 = np.matrix(c_price).reshape((length,1))
        X = np.hstack((one,x1,x2))
        return np.array(np.dot(np.dot(X.T,X).I,np.dot(X.T,Y))),now_price
    
    def optimize_price(self,competitor):
        argu,pj = self.ols(competitor)
        g,a,b = argu[0][0],-argu[1][0],argu[2][0]
        pi = (g+b*pj)/(2*a)
        if pi < 0:
            pi = 0
        self.update_price(pi)    
```

```python
def NASH(A,B):
    if A.model == 'linear' and B.model == 'linear':
        aprice = (2*A.G*B.alpha+A.beta*B.G)/(4*A.alpha*B.alpha-B.beta*A.beta)
        bprice = (2*B.G*A.alpha+B.beta*A.G)/(4*A.alpha*B.alpha-B.beta*A.beta)    
    elif A.model == 'linear' and B.model == 'exponential':
        aprice = (A.G + A.beta/B.alpha)/(2*A.alpha)
        bprice = 1/B.alpha
    elif A.model == 'linear' and B.model == 'logit':
        aprice = (A.G + A.beta/B.alpha)/(2*A.alpha)
        bprice = 1/B.alpha
    elif A.model == 'exponential' and B.model == 'linear':
        aprice = 1/A.alpha
        bprice = (B.G + B.beta/A,alpha)/(2*B.alpha)
    elif A.model == 'exponential' and B.model == 'exponential':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    elif A.model == 'exponential' and B.model == 'logit':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    elif A.model == 'logit' and B.model == 'linear':
        aprice = 1/A.alpha
        bprice = (B.G + B.beta/A,alpha)/(2*B.alpha)
    elif A.model == 'logit' and B.model == 'exponential':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    elif A.model == 'logit' and B.model == 'logit':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    aprofit = (A.G - A.alpha*aprice + A.beta*bprice)*aprice
    bprofit = (B.G - B.alpha*bprice + B.beta*bprice)*bprice 
    A.set_nash(aprice,aprofit)
    B.set_nash(bprice,bprofit)
```

```python
def graph(A,B):
    ap,ad,bp,bd,ar,br,cad,car,cbd,cbr = process(A,B)
    plt.plot(ap,color = 'limegreen',label = 'A')
    plt.plot(bp,color = 'indigo',label = 'B')
    plt.plot([0,len(ap)],[A.nash_price,A.nash_price],color = 'r',label = 'Astar',marker = 'o')
    plt.plot([0,len(bp)],[B.nash_price,B.nash_price],color = 'yellow',label = 'Bstar')
    plt.legend()
    plt.title('Current Price')
    plt.xlabel('Time')
    plt.ylabel('Current Price')
    plt.show()
def process(A,B):
    ap,ad,bp,bd = A.price,A.demand,B.price,B.demand
    ar = [ap[i]*ad[i] for i in range(len(ap))]
    br = [bp[i]*bd[i] for i in range(len(bp))]
    cad = list(np.cumsum(ad))
    car = list(np.cumsum(ar))
    cbd = list(np.cumsum(bd))
    cbr = list(np.cumsum(br))
    return ap,ad,bp,bd,ar,br,cad,car,cbd,cbr
```

```python
def initialize(A,B):
    NASH(A,B)
    A.update_demand(B.get_last_price())
    B.update_demand(A.get_last_price())
    A.update_price()
    B.update_price()
    A.update_demand(B.get_last_price())
    B.update_demand(A.get_last_price())
    return A,B    
```

```python
def action(A,B):
    A,B = initialize(A,B)
    while A.get_time() < 100:
        A.update_price()
        B.optimize_price(A)
        A.update_demand(B.get_last_price())
        B.update_demand(A.get_last_price())
        B.update_price()
        A.optimize_price(B)
        A.update_demand(B.get_last_price())
        B.update_demand(A.get_last_price())
    graph(A,B)
    def d(self,p):
        return max(0,self.last_price*(self.G - self.alpha*self.last_price + self.beta*p))
    def d1(self,p):
        return max(0,self.nash_price*(self.G - self.alpha*self.nash_price + self.beta*p))
    print(d(A,B.last_price))
    print(d(B,A.last_price))
    print(d1(A,B.nash_price))
    print(d1(B,A.nash_price))
```
