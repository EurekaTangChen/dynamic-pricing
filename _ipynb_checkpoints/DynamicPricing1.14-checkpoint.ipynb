```python
import numpy as np
import pandas as pd
import random
import math
import matplotlib.pyplot as plt
```

```python
class enterprise:
    def __init__(self,G,alpha,beta,n = 20,p = 5,
                 std = 0.5,rho = 0.1,model = 'linear'):
        self.model = model
        self.G = G
        self.alpha = alpha
        self.beta = beta
        self.history_number = n
        self.last_price = p
        self.std = std
        self.rho = rho
        self.price = [p]
        self.demand = []
        self.history_G = []
        self.history_alpha = []
        self.history_beta = []
        self.time = 1
        self.nash_price = 0
        self.nash_profit = 0
        
    def set_nash(self,price,profit):
        self.nash_price = price
        self.nash_profit = profit
    
    def get_time(self):
        return self.time
        
    def get_last_price(self):
        return self.last_price
    
    def return_price(self):
        return self.price
    
    def return_demand(self):
        return self.demand
        
    def check(self):
        if self.nash_profit <= 0:
            print("The nash equilibrium revenue is below 0!")
    
    def update_time(self):
        self.time += 1
    
    def update_price(self,new = None):
        if new == None:
            self.price.append(self.price[-1]+self.rho*self.time**(-0.25))
            self.last_price = self.price[-1]
        else:
            self.price.append(new)
            self.last_price = new
            
    def init_update_price(self,new = None):
        ud = 1 if random.random() > 0.5 else -1
        if new == None:
            self.price.append(self.price[-1]+ud*self.rho*self.time**(-0.25))
            self.last_price = self.price[-1]
        else:
            self.price.append(new)
            self.last_price = new
            
    def update_demand(self,p_another):
        new = self.get_demand(p_another)
        self.demand.append(new)
        self.update_time()
                  
    def get_demand(self,p_another):
        var = random.normalvariate(0,self.std)
        temp = self.G - self.alpha*self.last_price + self.beta*p_another
        if self.model == 'linear':
            demand = temp
        elif self.model == 'exponential':
            demand = math.exp(temp)
        elif self.model == 'logit':
            demand = math.exp(temp)/(1+math.exp(temp))
        else:
            print("The demand model you input has to be 'linear','exponential' or 'logit'!")
        if demand+var < 0:
            return 0
        else:
            return demand+var
    
    def ols(self,competitor):
        c_price = competitor.return_price()
        now_price = c_price[-1]
        c_price = c_price[:-1]
        length = len(c_price)
        if length > self.history_number:
            length = self.history_number
        Y = np.array(self.demand[-length:]).reshape((length,1))
        one = np.ones((length,1))
        x1 = np.matrix(self.price[-length:]).reshape((length,1))
        x2 = np.matrix(c_price[-length:]).reshape((length,1))
        X = np.hstack((one,x1,x2))
        return np.array(np.dot(np.dot(X.T,X).I,np.dot(X.T,Y))),now_price
    
    def optimize_price(self,competitor):
        argu,pj = self.ols(competitor)
        g,a,b = argu[0][0],-argu[1][0],argu[2][0]
        self.history_G.append(g)
        self.history_alpha.append(a)
        self.history_beta.append(b)
        pi = (g+b*pj)/(2*a)#加一个随机项
        if pi < 0:
            pi = 0
        self.update_price(pi)  
    
    def convergence(self,window=100):
        n = len(self.price)//window 
        nash = np.array(self.nash_price)
        RMSD = []
        RSSD = []
        def calculateRMSD(data):
            mean = np.mean(data)
            data = (data-mean)**2
            MSD = np.sum(data)/(window-1)
            rMSD = np.sqrt(MSD)
            return rMSD
        def calculateRSSD(data,nash):
            data = (data-nash)**2
            SSD = np.sum(data)/(window-1)
            rSSD = np.sqrt(SSD)
            return rSSD
        for i in range(n):
            data = self.price[i*window:(i+1)*window]
            RMSD.append(calculateRMSD(data))
            RSSD.append(calculateRSSD(data,nash))
        plt.figure(dpi=300)
        plt.subplot(1, 2, 1)
        plt.plot(RMSD,color = 'turquoise',linewidth = 0.8)
        plt.plot([0,n],[0,0],color = 'blueviolet',linewidth = 0.8)
        plt.title('RMSD')
        plt.subplot(1, 2, 2)
        plt.plot(RSSD,color = 'turquoise',linewidth = 0.8)
        plt.plot([0,n],[0,0],color = 'blueviolet',linewidth = 0.8)
        plt.title('RSSD')
        plt.tight_layout()

class smart_enterprise(enterprise):
    def __init__(self,G,alpha,beta,n = 20,p = 5,
                 std = 0.5,rho = 0.1,model = 'linear'):
        super().__init__(G,alpha,beta,n,p,std,rho,model)
    
    def optimize_price(self,competitor):
        argu,pj = self.ols(competitor)
        g,a,b = self.G,self.alpha,self.beta
        self.history_G.append(g)
        self.history_alpha.append(a)
        self.history_beta.append(b)
        pi = (g+b*pj)/(2*a)
        if pi < 0:
            pi = 0
        self.update_price(pi)
        
def NASH(A,B):
    if A.model == 'linear' and B.model == 'linear':
        aprice = (2*A.G*B.alpha+A.beta*B.G)/(4*A.alpha*B.alpha-B.beta*A.beta)
        bprice = (2*B.G*A.alpha+B.beta*A.G)/(4*A.alpha*B.alpha-B.beta*A.beta)    
    elif A.model == 'linear' and B.model == 'exponential':
        aprice = (A.G + A.beta/B.alpha)/(2*A.alpha)
        bprice = 1/B.alpha
    elif A.model == 'linear' and B.model == 'logit':
        aprice = (A.G + A.beta/B.alpha)/(2*A.alpha)
        bprice = 1/B.alpha
    elif A.model == 'exponential' and B.model == 'linear':
        aprice = 1/A.alpha
        bprice = (B.G + B.beta/A,alpha)/(2*B.alpha)
    elif A.model == 'exponential' and B.model == 'exponential':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    elif A.model == 'exponential' and B.model == 'logit':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    elif A.model == 'logit' and B.model == 'linear':
        aprice = 1/A.alpha
        bprice = (B.G + B.beta/A,alpha)/(2*B.alpha)
    elif A.model == 'logit' and B.model == 'exponential':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    elif A.model == 'logit' and B.model == 'logit':
        aprice = 1/A.alpha
        bprice = 1/B.alpha
    aprofit = (A.G - A.alpha*aprice + A.beta*bprice)*aprice
    bprofit = (B.G - B.alpha*bprice + B.beta*aprice)*bprice 
    A.set_nash(aprice,aprofit)
    B.set_nash(bprice,bprofit)

def graph(A,B,plot):
    if plot == "simple":
        graph1(A,B)
    elif plot == "complex":
        graph2(A,B)

def graph1(A,B):
    df1,df2 = process(A,B)
    plt.figure(figsize=(10,6),dpi=300)
    plt.subplot(2, 1, 1)
    plt.plot(df1.A_price,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df1.B_price,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([0,len(df1.A_price)],[A.nash_price,A.nash_price],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([0,len(df1.A_price)],[B.nash_price,B.nash_price],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History Price')
    plt.xlabel('Time')
    plt.ylabel('Current Price')
    plt.legend()
    
    plt.subplot(2, 1, 2)
    plt.plot(df1.A_revenue,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df1.B_revenue,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([0,len(df1.A_price)],[A.nash_profit,A.nash_profit],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([0,len(df1.A_price)],[B.nash_profit,B.nash_profit],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History Revenue')
    plt.xlabel('Time')
    plt.ylabel('Current Revenue')
    plt.legend()
    plt.tight_layout()

def graph2(A,B):
    df1,df2 = process(A,B)
    plt.figure(figsize=(10,20),dpi=300)
    plt.subplot(6, 1, 1)
    plt.plot(df1.A_price,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df1.B_price,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([0,len(df1.A_price)],[A.nash_price,A.nash_price],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([0,len(df1.A_price)],[B.nash_price,B.nash_price],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History Price')
    plt.xlabel('Time')
    plt.ylabel('Current Price')
    plt.legend()
    
    plt.subplot(6, 1, 2)
    plt.plot(df1.A_revenue,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df1.B_revenue,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([0,len(df1.A_price)],[A.nash_profit,A.nash_profit],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([0,len(df1.A_price)],[B.nash_profit,B.nash_profit],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History Revenue')
    plt.xlabel('Time')
    plt.ylabel('Current Revenue')
    plt.legend()
    
    plt.subplot(6, 1, 3)
    plt.plot(df1.A_cum_revenue,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df1.B_cum_revenue,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([1,len(df1.A_price)],[A.nash_profit,A.nash_profit*len(df1.A_price)],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([1,len(df1.A_price)],[B.nash_profit,B.nash_profit*len(df1.B_price)],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('Cumulative Revenue')
    plt.xlabel('Time')
    plt.ylabel('Cumulative Revenue')
    plt.legend()
    
    plt.subplot(6, 1, 4)
    plt.plot(df2.A_G,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df2.B_G,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([1,len(df2.A_G)],[A.G,A.G],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([1,len(df2.A_G)],[B.G,B.G],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History G')
    plt.xlabel('Time')
    plt.ylabel('History G')
    plt.legend()
    
    plt.subplot(6, 1, 5)
    plt.plot(df2.A_alpha,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df2.B_alpha,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([1,len(df2.A_G)],[A.alpha,A.alpha],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([1,len(df2.A_G)],[B.alpha,B.alpha],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History Alpha')
    plt.xlabel('Time')
    plt.ylabel('History Alpha')
    plt.legend()
    
    plt.subplot(6, 1, 6)
    plt.plot(df2.A_beta,color = 'turquoise',label = 'A',linewidth = 0.8)
    plt.plot(df2.B_beta,color = 'blueviolet',label = 'B',linewidth = 0.8)
    plt.plot([1,len(df2.A_G)],[A.beta,A.beta],color = 'crimson',label = 'Astar',linestyle = '--',linewidth=0.5)
    plt.plot([1,len(df2.A_G)],[B.beta,B.beta],color = 'darkorange',label = 'Bstar',linestyle = '--',linewidth=0.5)
    plt.title('History Beta')
    plt.xlabel('Time')
    plt.ylabel('History Beta')
    plt.legend()
    plt.tight_layout()
    
def process(A,B):
    dict1 = {'A_price':A.price,'B_price':B.price,
             'A_demand':A.demand,'B_demand':B.demand}
    df1 = pd.DataFrame(dict1)
    df1['A_revenue'] = df1['A_price']*df1['A_demand']
    df1['B_revenue'] = df1['B_price']*df1['B_demand']
    df1['A_cum_revenue'] = np.cumsum(df1['A_revenue'])
    df1['B_cum_revenue'] = np.cumsum(df1['B_revenue'])
    dict2 ={'A_G':A.history_G,'B_G':B.history_G,
            'A_alpha':A.history_alpha,'B_alpha':B.history_alpha,
            'A_beta':A.history_beta,'B_beta':B.history_beta}
    df2 = pd.DataFrame(dict2)
    return df1,df2

def initialize(A,B,n):
    NASH(A,B)
    A.check()
    B.check()
    A.update_demand(B.get_last_price())
    B.update_demand(A.get_last_price())
    def cold(A,B):
        A.init_update_price()
        B.init_update_price()
        A.update_demand(B.get_last_price())
        B.update_demand(A.get_last_price())
    for i in range(n):
        cold(A,B)
    return A,B   

def frame(A,B):
    pA = np.array(A.price)
    pB = np.array(B.price)
    dA = np.array(A.demand)
    dB = np.array(B.demand)
    rA = pA*dA
    rB = pB*dB
    trA = (A.G - A.alpha*pA +A.beta*pB)*pA
    trB = (B.G - B.alpha*pB +B.beta*pA)*pB
    trA[trA<0]=0
    trB[trB<0]=0
    df = pd.DataFrame({'pA':pA,'pB':pB,
                      'rA':rA,'rB':rB,'trA':trA,'trB':trB})
    return df
    
def action(A,B,Total = 200,plot = 'simple',window = 100,n=3,headline=10):
    A,B = initialize(A,B,n)
    while A.get_time() < Total:
        A.update_price()
        B.optimize_price(A)
        A.update_demand(B.get_last_price())
        B.update_demand(A.get_last_price())
        B.update_price()
        A.optimize_price(B)
        A.update_demand(B.get_last_price())
        B.update_demand(A.get_last_price())
    graph(A,B,plot)
    df = frame(A,B)
    A.convergence(window)
    B.convergence(window)
    print(df.head(headline))
    return df
```
